```python
#replaced two type as they are to be of same name
sui.replace('Bankruptcy or sudden change in economic','Bankruptcy or sudden change in economic status', inplace=True)
sui.replace('By other means(Please specify)', 'By other means', inplace=True)
```

```python
sui[sui['Total']==0] #where the suicide total is zero 
```

```python
sui=sui.drop(sui[sui.Total==0].index) #drop all rows where total is zero
```

```python
sui[sui['Total']==0]  #total with zeros thrown out
```

```python
sui.replace('A & N Islands(Ut)','A & N Islands', inplace=True) #total contains three types as with ut, all india so these are from the union territory part
sui.replace('Chandigarh(Ut)','Chandigarh', inplace=True)
sui.replace('D & N Haveli(Ut)','D & N Haveli', inplace=True)
sui.replace('Daman & Diu(Ut)','Daman & Diu', inplace=True)
sui.replace('Lakshadweep(Ut)','Lakshadweep', inplace=True)
sui.replace('Delhi(Ut)','Delhi', inplace=True)
```

```python
sui = sui.drop(sui[(sui.State == 'Total(Uts)')].index) #these come down states so
sui = sui.drop(sui[sui.State == 'Total(All India)'].index)
sui = sui.drop(sui[sui.State == 'Total(States)'].index)
```
